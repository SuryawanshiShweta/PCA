```python
df_new = df.values
```

```python
df_new
```

```python
from sklearn.preprocessing import scale
```

```python
df_normal = scale(df_new)
```

```python
df_normal
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
pca_values = pca.fit_transform(df_normal)
```

```python
pca_values
```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
# plot between PCA1 and PCA2 
x = pca_values[:,0:1]
y = pca_values[:,1:2]
z = pca_values[:2:3]
plt.scatter(x,y)
```

KMEANS

```python
# Importing library for KMeans
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(df_normal)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(4, random_state=30)
clusters_new.fit(df_normal)
```

```python
clusters_new.labels_
```

```python
kmean_cluster = clusters_new.labels_
```

```python
#Assign clusters to the data set
df['clusterid_new'] = clusters_new.labels_
```

```python
df.head()
```

```python
plt.scatter(df_new[kmean_cluster==0,0],df_new[kmean_cluster==0,1],s=50,marker='o',color='red')
plt.scatter(df_new[kmean_cluster==1,0],df_new[kmean_cluster==1,1],s=50,marker='o',color='blue')
plt.scatter(df_new[kmean_cluster==2,0],df_new[kmean_cluster==2,1],s=50,marker='o',color='green')
plt.scatter(df_new[kmean_cluster==3,0],df_new[kmean_cluster==3,1],s=50,marker='o',color='yellow')
plt.show()
```

Data Counts in Clusters

```python
plt.figure(figsize=(8,5))
df.groupby(df['clusterid_new']).count()['Type'].plot(kind='bar')
plt.xlabel('Cluster Numbers',size=12)
plt.ylabel('Type',size=12)
plt.title('Data counts in cluster',size=20)
```

Insights :

-  Here we are used KMEAN clustering from group purpose i.e. grouping the similers data points together but in KMEANS similirity is measured by using Eucliden Distance.

-  we can use this types of cluster for filling NA values, filtering the null values and after that checking the near clusters and according to that taking mean/mod/median we can fill our NA data more accureatly while filling randomly mean/mode/median of all data.

-  Creating Dendrogram and dendrogram is a diagram that shows hierarchical relationship between objects.

-  The main use of a dendrogram is to work out the best way to allocate objects to clusters

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df, method='single'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='average')
```
